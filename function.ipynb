```python
def mul():
    x=int(input('x = '))
    y=int(input('y = '))
    result = x*y
    print(result)
    return result

def mul2():
    total=mul()
    z=int(input("z = "))
    tot=total*z
    print(tot)
```

```python
def sum(in_list):
    cnt=0
    for x in in_list:
        cnt += int(x)
    return cnt
```

```python
in_list=input().split()
sum(in_list)
```

### lamda anonymous function

```python
def sum(x,y):
    return x+y
sum(4,5)
    
```

```python
f=lambda x,y: x+y
f(3,3)
```

Lamda in higher order

```python
def operation_higher(x,y,operation):
    return operation(x,y)
operation_higher(2,3,lambda x,y:x+y)
```

### map()

```python
num=[1,2,3,4]
num=list(map(lambda x:x**2,num))
num
```

### filter()

```python
num=list(filter(lambda num:num%2==1,num))
num
```

```python
from functools import reduce
reduce(lambda a,b:a+b,num)
```
