### Conditional statement

```python
x=30
y=40
if x<y:
    print('Y is greater than x')
elif(x==y)
    print('x==y')
else: 
    print('x is grater')
```

```python
num=float(input('Enter any number '))
if num%2==0:
    if num%5==0:
        print('2&5')
    else:
        print('NO 1')
else:
    print('No 2')
```

```python
bill = int(input('Enter bill '))
if bill<=100:
    print('NO charge')
elif bill<=200:
    print(bill*4)
else:
    print(bill*10)
```
