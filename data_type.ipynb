## Binary type

```python
byte_arr=bytearray('ABCDEF','utf-8')
type(byte_arr)
```

What is utf-8?
UTF-8 (Unicode Transformation Format - 8-bit) is a character encoding system.

To understand why it is necessary here, you have to remember one fundamental rule about computers: they do not understand letters or text; they only understand numbers (bytes).

When you try to turn a string of text (like 'ABCDEF') into a bytearray (which is strictly a list of numbers from 0 to 255), Python needs a "translator" or a "rulebook" to know exactly which numbers to assign to those letters. UTF-8 is that rulebook.
Here is what happens under the hood when you use it:

Python looks at the first letter, 'A'.

It checks the UTF-8 rulebook and sees that 'A' translates to the number 65.

It looks at 'B' and translates it to 66.

It does this for the whole string.
